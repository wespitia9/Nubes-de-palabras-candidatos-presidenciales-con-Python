# ¿Qué son las nubes de palabras?

Cuanto más grande y audaz aparece la palabra, más a menudo se menciona con un texto dado y más importante es. # **Nube de palabras, Gustavo Petro**

```python
# Importar BeautifulSoup para información de estructura HTML de nuestra solicitud
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url.text, 'lxml') # Extrayendo los datos de website_url
data = [element.text for element in soup.find_all('p')] # Recuperando texto de los fondos 'p'
data = str(data) # Conversión de datos a STR
```

```python
from stop_words import get_stop_words # Las palabras que queremos ignorar

stop_words_es = get_stop_words('es') # Palabras vacías en español
```

```python
from wordcloud import WordCloud, STOPWORDS
print('WordCloud está instalado y importado!')
```

```python
stop_words = ['año', 'mil', 'millones', 'pesos', 'ciento', 'El', 'En', 'NAceptar', 'n','n\'','xa0\'', 'tiempo', 'TIEMPO', 'CLARO'] + stop_words_es # Adición de palabras vacías adicionales
# Iniciar objeto de nube mundial
amlo_wc = WordCloud(
        background_color = 'white', #Fondo del tablero
        max_words = 1000, #Cantidad de palabras a visualizar
        stopwords = stop_words)
# Generate the word Cloud
amlo_wc.generate(data)
print('¡Se generan datos!')
```

```python
# Mostrar la nube de palabras]
plt.imshow(amlo_wc, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

```python
# regenerar la nube de palabras
amlo_wc.generate(data)

# Mostrar la nube
fig = plt.figure()
fig.set_figwidth(14) # Establecer ancho
fig.set_figheight(18) # Establecer altura

plt.imshow(amlo_wc, interpolation = 'bilinear')
plt.axis('off')
plt.show()
print('Nube de palabras de Gustavo Petro')
```
